"Loves me, loves me not" is a traditional game in which a person plucks off all the petals of a flower one by one, saying the phrase "Loves me" and "Loves me not" when determining whether the one that they love, loves them back. Given a number of petals, return a string which repeats the phrases "Loves me" and "Loves me not" for every alternating petal, and return the last phrase in all caps. Remember to put a comma and space between phrases.

INPUT : 3 ➞ "Loves me, Loves me not, LOVES ME"

INPUT : 6 ➞ "Loves me, Loves me not, Loves me, Loves me not, Loves me, LOVES ME NOT"

INPUT : 1 ➞ "LOVES ME"

```python
number = int(input(''))
output = ''
counter = 1
while counter <= number :
  
  if counter == number and counter % 2 != 0 :
    output += 'LOVES ME'
    break
   
  elif counter == number and counter % 2 == 0 :
    output += 'LOVES ME NOT'
    break
  if counter % 2 != 0 :
    output += 'Loves me, '
  else:
    output += 'Loves me not, '
  
  counter += 1
  
print(output)
    
```

```python
Group Anagrams

Title: Given an array of strings, you can combine words with different letters. Alphabetic words refer to strings with the same letters but arranged differently.

Given an array of strings, group anagrams together.

Example:

Input: ["eat", "tea", "tan", "ate", "nat", "bat"]

Output:

**[

["ate","eat","tea"],

["nat","tan"],

["bat"]

]

Explain:

All inputs are lowercase. Do not consider the order of the answer output. Note:

All inputs will be in lowercase. The order of your output does not matter.
```

```python
strs = ["eat", "tea", "tan", "ate", "nat", "bat", "cat", "tac"]
group = {}

for i in strs :
  kelime = "".join(sorted(i))
  if kelime in group :
    group[kelime].append(i)
  else :
    group[kelime] = [i]

print(list(group.values()))

```

```python
number = int(input(''))
fctrl = 1
if number == 1 :
  fctrl = 1
else:
  for i in range(1, number + 1) :
    fctrl = fctrl * i
print(fctrl)
```

```python
def for_loop() :
  result = []
  for i in range(1000000) :
    result.append(i)
  return result
```

```python
def list_comp() :
  return [i for i in range(1000000)]
```

Exercise-39  "Loves me, loves me not" is a traditional game in which a person plucks off all the petals of a flower one by one, saying the phrase "Loves me" and "Loves me not" when determining whether the one that they love, loves them back.
Given a number of petals, return a string which repeats the phrases "Loves me" and "Loves me not" for every alternating petal, and return the last phrase in all caps. Remember to put a comma and space between phrases

```python
n = int(input(''))
output = ''
for i in range (1, n+1) :
  if i % 2 != 0 :
    if i == n :
      output += 'LOVES ME'
    else :
      output += 'Loves Me, '
  else :
    if i == n :
      output += 'LOVES ME NOT'
    else:
      output += 'Loves Me Not, '
output

```
